```python

# Assuming you have two data frames df1 and df2

# Extract the columns you want to combine
column_from_yahoo_data1 = yahoo_data1['Adj Close PHNX.L']
column_from_yahoo_data2 = yahoo_data2['Adj Close LGEN.L']
column_from_yahoo_data3 = yahoo_data3['Adj Close AV.L']
column_from_yahoo_data4 = yahoo_data4['Adj Close MNG.L']
column_from_yahoo_data5 = yahoo_data5['Adj Close JUST.L']
column_from_yahoo_data6 = yahoo_data6['Adj Close BLND.L']
column_from_yahoo_data7 = yahoo_data7['Adj Close LAND.L']
column_from_yahoo_data8 = yahoo_data8['Adj Close GPE.L']
column_from_df = df['value']
# Combine the columns horizontally
combined_df = pd.concat([column_from_yahoo_data1,column_from_yahoo_data2,column_from_yahoo_data3,column_from_yahoo_data4,column_from_yahoo_data5,column_from_yahoo_data6,column_from_yahoo_data7,column_from_yahoo_data8,column_from_df], axis=1)
columns_to_exclude = ['value']  # Add your columns to exclude here

# Calculate rate of change for all columns except excluded ones
columns_to_change = combined_df.columns.difference(columns_to_exclude)
combined_df[columns_to_change] = combined_df[columns_to_change].apply(pd.to_numeric, errors='coerce')
combined_df[columns_to_change] = combined_df[columns_to_change].pct_change()

# Replace NaN values in selected columns with 0
combined_df[columns_to_change].fillna(0, inplace=True)
combined_df.fillna(0, inplace=True)
# Now you have the combined DataFrame
print(combined_df)
```

```python
# Assuming you have combined_df DataFrame and dropped NaN values

# Perform ADF test for each column
for column_name in combined_df.columns:
    result = adfuller(combined_df[column_name], autolag='AIC')
    
    # Extract results
    adf_statistic = result[0]
    p_value = result[1]
    critical_values = result[4]

    # Print results
    print(f'ADF Statistic for {column_name}: {adf_statistic}')
    print(f'p-value for {column_name}: {p_value}')
    print('Critical Values:')
    for key, value in critical_values.items():
        print(f'   {key}: {value}')
    print('\n')
```

```python
# Compute correlation matrix
correlation_matrix = combined_df.corr()

# Print correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Optional: Visualize correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Convert all columns to numeric (excluding non-numeric columns)
combined_df_numeric = combined_df.apply(pd.to_numeric, errors='coerce')

# Drop rows with missing values
combined_df_numeric.dropna(inplace=True)

# Define independent variables (excluding 'value')
independent_vars = combined_df_numeric.drop(columns=['value'])

# Add constant term to independent variables
independent_vars = sm.add_constant(independent_vars)

# Define dependent variable
dependent_var = combined_df_numeric['value']

# Fit the linear regression model
model = sm.OLS(dependent_var, independent_vars).fit()

# Print model summary
print(model.summary())
```

All coefficients for the independent variables, except for "Adj Close LAND.L," are not statistically significant (p-values > 0.05). This indicates that changes in the adjusted prices of PHNX.L, LGEN.L, AV.L, MNG.L, JUST.L, BLND.L, and GPE.L stocks do not have a significant relationship with the dependent variable. However, "Adj Close LAND.L" has a statistically significant coefficient (p-value = 0.009), suggesting that changes in the adjusted price of LAND.L stock may have a significant impact on the dependent variable. Further investigation into this particular relationship is warranted to understand its implications.







```python
# Plot observed vs. predicted values
plt.figure(figsize=(10, 6))
plt.scatter(model.fittedvalues, dependent_var, color='blue', label='Observed vs. Predicted')
plt.plot(dependent_var, dependent_var, color='red', label='Perfect Fit')  # Perfect fit line (y=x)
plt.title('Observed vs. Predicted Values')
plt.xlabel('Predicted Values')
plt.ylabel('Observed Values')
plt.legend()
plt.grid(True)
plt.show()
```
